```python
an_array=np.array([[11,12,13],[21,22,23],[31,32,33],[41,42,43]])
print(an_array)
col_indices=np.array([0,1,2,0])
row_indices=np.arange(4)
print(row_indices)
print(col_indices)

for row,col in zip(row_indices, col_indices):
    print(row,",",col)
an_array[row_indices, col_indices]+=100000
print(an_array)
```

# set

```python
s1=np.array(['desk', 'chair', 'bulb'])
s2=np.array(['lamp', 'bulb', 'chair'])
print(np.intersect1d(s1,s2))
print(np.union1d(s1, s2))
print(np.setdiff1d(s1,s2))
print(np.setdiff1d(s2,s1))
```

### In
### s2의 요소가 s1의 요소 안에 있으면 true

```python
print(np.in1d(s1,s2))
```

# matplotlib

```python
apple_stock_price = [93.95, 112.15, 104.05, 144.85, 169.49]
ms_stock_price=[39.01, 50.29, 57.05, 69.98, 94.39]
year=[2014, 2015, 2016, 2017, 2018]
plt.plot(year, apple_stock_price)
plt.show()
plt.plot(year, ms_stock_price)
plt.show()
plt.plot(year,apple_stock_price, year,ms_stock_price)
plt.show()
```

## chart Labelling

```python
plt.plot(year,apple_stock_price)
plt.title('Stock Price Comparision')
plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.show()
```

```python
plt.plot(year,apple_stock_price,year,ms_stock_price)
plt.title('Stock Price Comparison')
plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.legend(['apple stock', 'ms stock'])
plt.show()
```

```python
plt.scatter(year, apple_stock_price)
```

```python
plt.plot(year, apple_stock_price, 'r--', year, ms_stock_price, 'g')
```
